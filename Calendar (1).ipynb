```python
def leapyear(a) :
    if a%4 != 0:
        return False
    elif (a%4 == 0) and (a%100 != 0):
        return True
    elif (a%100 == 0) and (a%400 != 0):
        return False
    elif (a%400 == 0) and (a%3200 !=0):
        return True
    else :
        return False
```

```python
def dayOfWeek (date,month,year):
    d = date
    m = month
    y = year
    if m <=2:
        m += 12
        y -= 1
    k = y%100
    j = y//100
    #equation for the day of the week
    h = (d + (13*(m+1)//5) + k + (k//4) + (j//4) - 2*j) %7
    #equation for coverting into ISO day of the week
    d = ((h+5) %7)+1
    #d-1 for array as array starts at 0
    d = int(d-1)  
    weeks = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
    d = Weeks[d]
    return (d)
```
